```python
import pandas as pd

from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://github.com/miningAH"
```

```python
headers = requests.utils.default_headers()

headers.update(
    {
        'User-Agent': 'My User Agent 1.0',
    }
)

page = requests.get(url, headers=headers)
```

```python
page
```

```python
soup = BeautifulSoup(page.text,  'html.parser')
```

```python
print(soup)
```

```python
projectLinks = [];
for a in soup.findAll('a'):
  if '/miningAH/' in a['href']:
    projectLinks.append(a);
```

```python
projectLinks
```

```python
portfolioProjects=[];

for p in projectLinks:
  projects = p.get_text()
  portfolioProjects.append(projects)
```

```python
print(portfolioProjects)
```

```python
text_string = ", ".join(portfolioProjects)
text_string
```

```python
new = text_string.replace("\n", "").replace(" ", "").replace(",", ", ")
new
```

```python
# NOTE: the below variable needs to be adjusted everytime a new project is added
num_chars_to_remove = 148

my_projects = new[:-num_chars_to_remove]
my_projects
```

```python
my_projects_list = my_projects.split(',')
print("My current GitHub projects are as follows:")
#for p in my_projects_list:
 #   print(f"â€¢ {p}")

for index, item in enumerate(my_projects_list, start=1):
    print(f"{index}. {item}")
```
